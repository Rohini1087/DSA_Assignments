# Importing Pandas,numpy,matplotlib,and seaborn Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Read the Iris dataset to python environment.

```python
iris_data=pd.read_excel('iris.xls')
iris_data
```

# 2. Display the columns in the dataset.

```python
iris_data.columns
```

# 3. Calculate the mean of each column of the dataset.

```python
iris_data[['SL','SW','PL','PW']].mean()
```

# MEAN of the dataset based on Classification

```python
iris_data.groupby('Classification').mean()
```

# 4. Check for the null values present in the dataset

```python
iris_data.isnull().values.any()
```

# To fetch the row if there is any null value

```python
iris_nan  = iris_data[iris_data.isnull().any(axis=1)]
```

```python
iris_nan
```

# 5. Perform meaningful visualizations using the dataset

# 1. Scatter Plot of Classification vs Sepal Length

```python
plt.figure(figsize=(8,6))
plt.scatter(iris_data['Classification'],iris_data['SL'],s=12,marker='^')
plt.title('Plot of Classification Vs Sepal Length',size=16)
plt.xlabel('Classification')
plt.ylabel('Sepal Length')
plt.tight_layout()
plt.grid()
```

# 2. Boxplot of Classification vs PetalLength

```python
sns.boxplot(x='Classification',y='PL',data=iris_data).set(title='Plot of Classification Vs Petal Length')
```

# 3. ViolinPlot of Classification vs Petal Width

```python
sns.violinplot(x='Classification',y='PW',data=iris_data).set(title='Plot of Classification Vs Petal Width')
```

# 4.Strip Plot of Classification vs Sepal Width

```python
sns.stripplot(x='SW',y='Classification',data=iris_data).set(title='Plot of Classification Vs Sepal Width')
```

# 5.Scatterplot of SepalLength and sepal Width based on Classification

```python
# Scatterplot using seaborn library
plt.figure(figsize=(10,3))
sns.set_style("whitegrid")
sns.scatterplot(x='SL', y='SW',data=iris_data, hue='Classification').set(title='Plot of SL Vs SW based on Classification')
```

# 6.Scatter plot of PL and PW based on Classification

```python
# Scatterplot using seaborn library
plt.figure(figsize=(10,3))
sns.set_style("whitegrid")
sns.scatterplot(x='PL', y='PW',data=iris_data, hue='Classification').set(title='Plot of PL Vs PW based on Classification')
```
